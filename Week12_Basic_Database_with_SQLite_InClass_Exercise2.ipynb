- <font size=3>Use the previous SQLite sample database, HospitalDB.db, perform the following tasks</font>
    - <font size=3>Prompt the user to provide a Hospital ID and then use a placeholder to retrieve the details of the hospital corresponding to the given Hospital ID</font>
    - <font size=3>Prompt the user to input a Doctor ID and then utilise a placeholder to retrieve the details of the doctor associated with the provided Doctor ID</font>
    - <font size=3>Prompt the user to input a specialty and salary, then use a placeholder to retrieve the list of doctors based on the provided specialty and salary</font>

```python
#import sqlite module
import sqlite3

#connect python to SQLite database
conn = sqlite3.connect('HospitalDB.db')

#create a cursor object
cursor = conn.cursor()

# Prompt the user to provide a Hospital ID
user_input = input("Please enter a Hospital ID: ")

# Create an SQL query
sql_statement = "SELECT * FROM Hospitals WHERE Hospital_Id = ?"


# Execute SQL query
cursor.execute(sql_statement, (user_input,))

#fetch result from a SELECT statement
hospital_1 = cursor.fetchone()

# Display result from query
print(hospital_1)
print(f"Hospital ID: {hospital_1[0]}, Hospital Name: {hospital_1[1]}, Bed Count: {hospital_1[2]}, Address: {hospital_1[3]}")
```

```python
# Prompt the user to input a Doctor ID
user_input = input("Please enter a Doctor ID: ")

# Create an SQL query
sql_statement3 = "SELECT * FROM Doctors WHERE Doctor_Id = ?"

# Execute SQL query
cursor.execute(sql_statement3, (user_input,))

#fetch result from a SELECT statement
doctor_1 = cursor.fetchone()

# Display result from query 
print(doctor_1)
print(f"Doctor ID: {doctor_1[0]}, Doctor Name: {doctor_1[1]}, Hospital ID: {doctor_1[2]}, Joining Date: {doctor_1[3]}, Speciality: {doctor_1[4]}, Salary: {doctor_1[5]}, Experience: {doctor_1[6]}")
```

```python
# Prompt the user to input a specialty and salary
specialty = input("Please enter a specialty: ")
salary = input("Please enter a salary: ")

# Create an SQL query
sql_statement2 = "SELECT * FROM Doctors WHERE Speciality = ? AND Salary = ?"

# Execute SQL query
cursor.execute(sql_statement2, (specialty, salary))
#fetch result from a SELECT statement
doctor_2 = cursor.fetchall()


# Display result from query 
print(doctor_2)
```
